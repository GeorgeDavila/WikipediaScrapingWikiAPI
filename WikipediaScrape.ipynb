###Cleaning the List

Looking in the 'tests subsection above you can see that searching with and without parentheses lead to different results. The ones with parentheses are worse, so we will remove them. Lets write the .content string to a txt file so manage it more easily

```
rodent_str = rodents_wikiList.content
rodent_str
```

```
import os

os.system('touch list_of_rodents_wikicontent.txt')  #Creates the txt file in your local directory. If using colab you'll noticed it created a txt in your colab directory 

open('list_of_rodents_wikicontent.txt','w').write(rodent_str) #writes rodent_str into the txt file
#Should see that the txt file now contains the wiki list 
#CAREFUL to open the file as write, NOT append. Append will add a new copy text to this file every time you run it, ie youll get duplicates
#Although if working with a list of list like the aforemention Lists of animals page you might want to use append
```

```
# Now we split the lines in the list, ie of the txt file 

lines = open('list_of_rodents_wikicontent.txt').read().splitlines()   #Returns the lines as a list of the lines
print(lines)
init_numoflines = len(lines)
print('Number of lines =  ',  init_numoflines )


post_section = lines.index('== See also ==') - 1   #should be NUM line in txt file -2  #use .index() to find the == See also section ==, now we empty anything after it
print( 'post_section =  ' + str(post_section) )

first_empty = lines.index('')
print('FIRST EMPTY STRING APPEARS AT index =    ', first_empty) #Anything b4 first empty space is the content we want to remove 

for i in range( len(lines) ):
    if i > post_section:
      lines[i] = ''
    if i < first_empty:      # REMOVE DESCRIPTIVE CONTENT ie anything b4 list, can also do by re
      lines[i] = ''

print(lines)

#Turn any lines that contain '=' into empty string ''
for i in range( len(lines) ):
    if lines[i].find('=') != -1:
      lines[i] = ''
 
print(lines)

#Remove all empty lines #Note that whitespace is included in the txt file to begin with, which is why we dont intergrate this with the previou steps
while('' in lines) : 
    lines.remove('')


print(lines)

print('Number of lines of final array =  ',  len(lines), '           Difference in Num of lines between this array and the initial one =  ', init_numoflines - len(lines))

```

**Remove stuff in parentheses:**

Note how the list is structured such that some lines with parentheses also have a comma before it such as 'North African gundi, (C. gundi)'
Some lines have no such comma

so we'll be excising BOTH the structures  ', (....)' and '(....)' 
starting with the comma one then the parentheses structure

You dont have to remove parentheses for every list, play around with it yourself 

```
#Removing stuff in parentheses

excise_str1 = 'North African gundi, (C. gundi,)'
excise_start1 = excise_str1.find(',')
excise_end1 = excise_str1.rfind(')')

new_str1 = excise_str1[ : excise_start1 ]

print(excise_str1 )
print( new_str1 )
```

Does the trick, now iterate over the string with this method

Now have our cleaned up Array of terms to search over

```
for i in range( len(lines) ):  #recall that lines was restructured to be of different length, so we include the len() rather than any preset value
    if lines[i].find('(') != -1:
      lines[i] = lines[i][ : lines[i].find(',')]
    if lines[i].find('(') != -1:                #NOW we catch the lines that have parentheses but no comma 
      lines[i] = lines[i][ : lines[i].find('(')]

print(lines)

#idk if some of these terms contain extraneous commas. So to play it safe I only excise stuff b4 commas in items with parentheses
```

####List-specific cleaning 

Need to remove any lines that contain 'genus',  'subfamily', or 'family'  and those terms beginning in capital letters

```
#Turn any lines that contain 'genus',  'subfamily', or 'family' into empty string ''

for i in range( len(lines) ):
    if lines[i].find('Genus') != -1:    #NOTE we have to do the capital and lowercase versions 
      lines[i] = ''
    if lines[i].find('Subfamily') != -1:
      lines[i] = ''
    if lines[i].find('Family') != -1:
      lines[i] = ''
    if lines[i].find('genus') != -1:
      lines[i] = ''
    if lines[i].find('subfamily') != -1:
      lines[i] = ''
    if lines[i].find('family') != -1:
      lines[i] = ''

#Remove all empty lines #Note that whitespace is included in the txt file to begin with, which is why we dont intergrate this with the previou steps
while('' in lines) : 
    lines.remove('')


print('Number of lines of final array =  ',  len(lines), '           Difference in Num of lines between this array and the initial one =  ', init_numoflines - len(lines))
print('NOTE: WE RETRUCTURED IT AGAIN BE CAREFUL NOT TO USE LENGTH FROM EARLIER') 
```

Now do a wiki search over all those terms 

THIS wiki.page() FUNCTION WANTS TO BREAK AT EVERY LITTLE IMPASSE SO WE'LL DO A HAVE TO ACCOUNT FOR ERRORS 

```
import os 

lines_str = str(lines)
os.system('list_of_rodents_cleaned.txt')
open('list_of_rodents_cleaned.txt','w').write( lines_str )

os.system('touch output_text.txt')
```

```
#pageArray = [None] * len(lines) #if you can 

for i in range( len(lines) ):
    try:
        open('output_text.txt','a+').write( wiki.page( lines[i] , pageid=None, auto_suggest=True, redirect=True, preload=False).content )
    except:
        pass
```

The above cell writes the contents of each page to output_text.txt 

The API is a bit of a pain and will frequently produce errors when it cant find a page using Wikipedias in-built search. So we ignore these instance by layering the try: and except: pass parts so it skips over the error instances.
